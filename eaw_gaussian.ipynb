```python

from cpl import bi_gau_dist

L = np.array([[1, 0], 
              [2, 1]])
M = np.array([[1, 0], 
              [-0.5, 1]])
a = np.array([0,0])
b = np.array([6,-6])
A = L.dot(L.T)
B = M.dot(M.T)
print(A)
print(B)
# print(L)
# print(M)
print('LTM')
print(L.T.dot(M))


plot_scatter_interpolation(a, b, A, B, gau_cpl=bi_gau_dist, n_sample=1000, n_time=3)

X, Y = bi_gau_dist(a, b, A, B)
plt.title(r"Adapted 2-Wasserstein ($\rho = 1$)")
plt.show()

ns = 50
Xs = X[:ns]
Ys = Y[:ns]
Gs = np.diag(np.ones(len(Xs)))

ot.plot.plot2D_samples_mat(Xs, Ys, Gs, alpha=0.1, c="k")
plt.plot(X[:, 0], X[:, 1], "+b", label="Source samples")
plt.plot(Y[:, 0], Y[:, 1], "xg", label="Target samples")
plt.legend()
```
