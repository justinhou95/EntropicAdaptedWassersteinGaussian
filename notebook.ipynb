```python
%load_ext autoreload
%autoreload 2
import numpy as np
import seaborn as sns
sns.set_style()
from src.compare import compare_rho, compare_scatter_contor, compare_line
import matplotlib.pyplot as plt
```

```python
# When $L^T M$ has zero diagonal,
# AW2-optimizers may not be unique 
# AW2-optimizers may not be gaussian
# AW2-optimizers may not be Monge-type

L = np.array([[1, 0], 
              [2, 1]])
M = np.array([[1, 0], 
              [-0.5, 1]])
a = np.array([0, 0])
b = np.array([6, -6])
A = L.dot(L.T)
B = M.dot(M.T)
print('A:\n', A, '\nB:\n', B, '\nL:\n', L, '\nM:\n', M, '\nLTM:\n', L.T.dot(M))
compare_rho(a,b,A,B)
plt.savefig("assets/compare_rho.png", bbox_inches="tight")
```

```python
# When $L^T M$ have positive diagonal

L = np.array([[1, 0], 
              [2, 1]])
M = np.array([[2, 0], 
              [-0.5, 1]])
a = np.array([0, 0])
b = np.array([6, -6])
A = L.dot(L.T)
B = M.dot(M.T)
print('A:\n', A, '\nB:\n', B, '\nL:\n', L, '\nM:\n', M, '\nLTM:\n', L.T.dot(M))
compare_scatter_contor(a,b,A,B)
```

```python
# When $L^T M$ have positive/negative diagonal
# Gaussian may degenerate during McCann interpolation

a = np.array([0, 0])
b = np.array([6, -6])
L = np.array([[1, 0], 
              [2, 1]])
M = np.array([[1, 0], 
              [-1, 1]])
A = L.dot(L.T)
B = M.dot(M.T)
print('A:\n', A, '\nB:\n', B, '\nL:\n', L, '\nM:\n', M, '\nLTM:\n', L.T.dot(M))
compare_scatter_contor(a,b,A,B)
plt.savefig("assets/compare_scatter_contor.png", bbox_inches="tight")
```
